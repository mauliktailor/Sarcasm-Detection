```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import  pad_sequences
```

```
sentences = ['Will is a data scientist',
             'Adam is a data Analyst!!',
             'Ok what is up with adam, the data scientist',
             'come on']
```

```
tokenizer = Tokenizer(num_words=100, oov_token="<OOV>")
```

```
tokenizer.fit_on_texts(sentences)
```

```
word_index = tokenizer.word_index
```

```
word_index
```

```
test_sentence = ['This is will bhai up in adam come']
```

```
seq = tokenizer.texts_to_sequences(sentences)
print(seq)
```

##PADDING

```
pad_seq = pad_sequences(seq, padding='post', value=0) # padding can be pre and post and we can change the value
```

#Brown Corpus

```
import nltk
nltk.download('brown')
nltk.download('nonbreaking_prefixes')
nltk.download('perluniprops')

from nltk.corpus import brown
from nltk.tokenize.moses import MosesDetokenizer

mdetok = MosesDetokenizer()

brown_natural = [mdetok.detokenize(' '.join(sent).replace('``', '"').replace("''", '"').replace('`', "'").split(), return_str=True)  for sent in brown.sents()]
```

```
brown_natural[0]
```

```
tokenizer2 = Tokenizer(oov_token='<OOV>')
```

```
tokenizer2.fit_on_texts(brown_natural)
```

```
seq1 = tokenizer2.texts_to_sequences(brown_natural)
```

```
padded_seq1 = pad_sequences(seq1, padding='post')
```

```
padded_seq1.shape
```
